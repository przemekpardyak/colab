```python
import torch

print(torch.__version__)
print(torch.cuda.is_available())
print(torch.cuda.device_count())
print(torch.cuda.current_device())
print(torch.cuda.get_device_name(0))


```

```python
print(torch.tensor(1))
print(torch.tensor([1, 2]))
print(torch.tensor([[1, 2],[3, 4]]))
print(torch.tensor([[[1, 2],[3, 4]],[[11, 22],[33, 44]]]))

t0a = torch.tensor(1)
print(t0a.dtype)
t0b = torch.tensor(1.0)
print(t0b.dtype)

t3 = torch.tensor([[[1, 2, 3],[4, 5, 6]],[[11, 22, 33],[44, 55, 66]]])
print(t3)
print("shape:", t3.shape)
print("view:", t3.view(3, 2, 2))
print("reshape:", t3.reshape(3, 2, 2))
print("reshape:", t3.reshape(2, 3, 2))
print("reshape:", t3.reshape(12))

t2 = torch.tensor([[1, 2, 3],[4, 5, 6]])
print(t2)
print("T:", t2.T)
print("M:", t2.matmul(t2.T))
print("M:", t2 @ t2.T)

```
